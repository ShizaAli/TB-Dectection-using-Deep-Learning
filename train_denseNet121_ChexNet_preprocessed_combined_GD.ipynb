```
from keras.applications.densenet import DenseNet121
from keras.preprocessing import image
from keras.models import Model
from keras.layers import Dense, GlobalAveragePooling2D
from keras import backend as K
from keras.layers import Input
from keras.applications.densenet import preprocess_input, decode_predictions
```

```


# create the base pre-trained model
input_tensor = Input(shape=(256, 256, 3)) 
base_model = DenseNet121( input_tensor=input_tensor, include_top=False)
weights_path='/content/gdrive/My Drive/new/chexnet_weights.h5'
# weights_path='/content/gdrive/My Drive/SPROJ1/weights-82-0.2654-0.9000.hdf5'
base_model.load_weights(weights_path, by_name='True')
 

# add a global spatial average pooling layer
x = base_model.output
x = GlobalAveragePooling2D()(x)
# let's add a fully-connected layer
x = Dense(512, activation='relu')(x)
# let's add a fully-connected layer
x = Dense(256, activation='relu')(x)
# and a logistic layer -- let's say we have 200 classes
predictions = Dense(num_classes, activation='softmax')(x)

# this is the model we will train
model = Model(inputs=base_model.input, outputs=predictions)

# first: train only the top layers (which were randomly initialized)
# i.e. freeze all convolutional InceptionV3 layers
for layer in base_model.layers:
    layer.trainable = False

# for layer in model.layers[:NB_IV3_LAYERS_TO_FREEZE]:
#       layer.trainable = False
#    for layer in model.layers[NB_IV3_LAYERS_TO_FREEZE:]:
#       layer.trainable = True

# compile the model (should be done *after* setting layers to non-trainable)
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# # train the model on the new data for a few epochs
# # model.fit_generator(...)
```

```
from keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
image_size = 256
#Tried to ue 512, but unable to run that on colab
#exhaust resources error

nb_train_samples = 1049     # number of files in training set
batch_size = 16

EPOCHS = 100
STEPS = ceil(nb_train_samples / batch_size)

# ## Specify the values for all arguments to data_generator_with_aug.
train_datagen_with_aug = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    horizontal_flip= True,
    vertical_flip=True,
    rotation_range=180,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    fill_mode='nearest')

data_generator_no_aug = ImageDataGenerator(preprocessing_function=preprocess_input)


train_generator = train_datagen_with_aug.flow_from_directory(
       directory = train_dir,
       target_size = (image_size, image_size),
       batch_size = batch_size,
       class_mode = 'categorical')

validation_generator = data_generator_no_aug.flow_from_directory(
        shuffle=False,
       directory = val_dir,
       target_size = (image_size, image_size),
       class_mode = 'categorical')

test_generator = data_generator_no_aug.flow_from_directory(
       directory = test_dir,
       target_size = (image_size, image_size),
       batch_size = batch_size,
       class_mode = 'categorical',shuffle=False)
```

```
def loadmodel():
     
  # create the base pre-trained model
  input_tensor = Input(shape=(256, 256, 3)) 
  base_model = DenseNet121( input_tensor=input_tensor, include_top=False)

  x = base_model.output
  x = GlobalAveragePooling2D()(x)
  # let's add a fully-connected layer
  x = Dense(512, activation='relu')(x)
  # let's add a fully-connected layer
  x = Dense(256, activation='relu')(x)

  # and a logistic layer -- let's say we have 200 classes
  predictions = Dense(num_classes, activation='softmax')(x)

  # this is the model we will train
  model = Model(inputs=base_model.input, outputs=predictions)
  

  # compile the model (should be done *after* setting layers to non-trainable)
  model.compile(optimizer='adam',
                loss='categorical_crossentropy',
                metrics=['accuracy'])

  return model
```

```

import os

def traverse_dir(path_sub,tb_ntb):
  y_pred=[]
  y_true=[]

  path_main="/content/gdrive/My Drive/SPROJ1/all_data/"

  if tb_ntb==1:
    state=1
    path_tb_ntb="tb/"
  else:
    state=0
    path_tb_ntb="ntb/"
  
  path = path_sub + path_tb_ntb
  dir=os.path.join(path_main, path)

  total=0
  right=0
  wrong=0

  for filename in os.listdir(dir):
    if filename.endswith(".png"): 
        print(filename)
        img = image.load_img(dir + filename, target_size=(image_size, image_size))
        preds = predict(model, img)
        print(preds)
        if preds[0]>preds[1]:
          y_pred.append(0)
        else:
          y_pred.append(1)
        
        y_true.append(state)
        print(y_pred[total], y_true[total])
        if y_pred[total]!=y_true[total]:
          wrong+=1
        else:
          right+=1
        total+=1

        continue
    else:
        continue
  print(right,wrong,total, (right/total))
  return y_pred, y_true
```
